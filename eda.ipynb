```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x_scaled = scaler.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x_scaled, y, test_size=0.33, random_state=42)
```

```python
from sklearn.model_selection import GridSearchCV
import math
from sklearn.metrics import mean_squared_error
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
from sklearn.linear_model import Ridge
ridge =Ridge()

parameters = {'alpha':[1e-15,1e-3,1e-2,1,5,10,20,30,40,50]}
rr = GridSearchCV(ridge,parameters,scoring='neg_mean_squared_error',cv=5)
rr.fit(X_train,y_train)
pred2=rr.predict(X_test)
rmse2 = math.sqrt(mean_squared_error(y_test,pred2))

print("Rmse : ",rmse2)
```

```python
from sklearn.linear_model import Lasso
lasso =Lasso()

las = GridSearchCV(lasso,parameters,scoring='neg_mean_squared_error',cv=5)
las.fit(X_train,y_train)
pred2=las.predict(X_test)
rmse2 = math.sqrt(mean_squared_error(y_test,pred2))
print("Rmse : ",rmse2)
```

```python
from sklearn.linear_model import ElasticNet
elastic = ElasticNet()
elasto=GridSearchCV(elastic,parameters,scoring='neg_mean_squared_error',cv=5)
elasto.fit(X_train,y_train)
pred2=elasto.predict(X_test)
rmse2 = math.sqrt(mean_squared_error(y_test,pred2))

print("Rmse : ",rmse2)
```

```python
from sklearn.svm import SVR
svr = SVR()
svr.fit(X_train,y_train)
pred2=svr.predict(X_test)
rmse2 = math.sqrt(mean_squared_error(y_test,pred2))

print("Rmse : ",rmse2)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
from sklearn.ensemble import RandomForestRegressor
randomforest = RandomForestRegressor()
randomforest.fit(X_train,y_train)
pred2=randomforest.predict(X_test)
rmse2 = math.sqrt(mean_squared_error(y_test,pred2))

print("Rmse : ",rmse2)
```
